```python
import os
import tensorflow as tf
from PIL import Image
import numpy as np

def preprocess_multiple_folders(
    input_folders, 
    output_folder="/kaggle/working/tarot_resized", 
    target_size=128,
    padding_color=(0, 0, 0)
):
    """
    Resizes and pads images from multiple folders into a single output folder.

    Parameters:
        input_folders (list): List of folders containing raw images
        output_folder (str): Destination for resized & padded images
        target_size (int): Final width/height in pixels (e.g., 128)
        padding_color (tuple): RGB color used for padding
    """
    os.makedirs(output_folder, exist_ok=True)
    
    # Collect all image paths from all input folders
    all_image_paths = []
    for folder in input_folders:
        paths = [
            os.path.join(folder, fname)
            for fname in os.listdir(folder)
            if fname.lower().endswith(('.jpg', '.jpeg', '.png'))
        ]
        all_image_paths.extend(paths)
    
    print(f"🔍 Found {len(all_image_paths)} total images from {len(input_folders)} folders.")

    for img_path in all_image_paths:
        try:
            img = Image.open(img_path).convert("RGB")
            original_w, original_h = img.size

            # Calculate scale and resize
            scale = target_size / max(original_w, original_h)
            new_w = int(original_w * scale)
            new_h = int(original_h * scale)
            resized_img = img.resize((new_w, new_h), Image.BILINEAR)

            # Pad to square
            new_img = Image.new("RGB", (target_size, target_size), padding_color)
            paste_x = (target_size - new_w) // 2
            paste_y = (target_size - new_h) // 2
            new_img.paste(resized_img, (paste_x, paste_y))

            # Normalize and save
            np_img = np.asarray(new_img) / 255.0
            filename = os.path.splitext(os.path.basename(img_path))[0] + ".png"
            save_path = os.path.join(output_folder, filename)
            tf.keras.preprocessing.image.save_img(save_path, np_img)

        except Exception as e:
            print(f"⚠️ Failed to process {img_path}: {e}")

    print(f"✅ All images saved to {output_folder}")
```

```python
import tensorflow as tf

# Hyperparameters
IMG_SIZE    = TARGET_SIZE  # your padded size (e.g. 128)
BATCH_SIZE  = 32
AUTOTUNE    = tf.data.AUTOTUNE
DATA_GLOB   = "/kaggle/working/tarot_resized/*.png"
SPLIT_RATIO = 0.8          # 80% train / 20% val

# 1) List + shuffle file paths
all_files = tf.data.Dataset.list_files(DATA_GLOB, shuffle=True)

# 2) Compute split counts
total_count = len(tf.io.gfile.glob(DATA_GLOB))
train_count = int(total_count * SPLIT_RATIO)

# 3) Split into train / val
train_files = all_files.take(train_count)
val_files   = all_files.skip(train_count)

# 4) Common loader + preprocess
def load_and_preprocess(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_png(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)   # [0,1]
    # Safety resize (should already be correct shape)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    return img, img  # input=output for autoencoder

# 5) Build the datasets
train_ds = (
    train_files
      .map(load_and_preprocess, num_parallel_calls=AUTOTUNE)
      .batch(BATCH_SIZE)
      .prefetch(AUTOTUNE)
)

val_ds = (
    val_files
      .map(load_and_preprocess, num_parallel_calls=AUTOTUNE)
      .batch(BATCH_SIZE)
      .prefetch(AUTOTUNE)
)

# Now you can plug these directly into model.fit()
# autoencoder.fit(train_ds, validation_data=val_ds, epochs=…)
```

```python
import glob, os
import tensorflow as tf

# ── HYPERPARAMETERS ───────────────────────────────────────────────────────────
IMG_SIZE    = TARGET_SIZE        # padded square size
BATCH_SIZE  = 32
AUTOTUNE    = tf.data.AUTOTUNE
SPLIT_RATIO = 0.8        # train/val split
DATA_DIR    = "/kaggle/working/tarot_resized"  # where your images are saved

NUM_MASKS   = 10          # number of random masks to apply per image
MASK_SIZE   = IMG_SIZE // 10   # default mask size
UL_REGION   = IMG_SIZE // 10   # "no-mask" upper-left region
# ──────────────────────────────────────────────────────────────────────────────

# 1) Gather image file paths
extensions = ("*.png", "*.jpg", "*.jpeg")
all_files = []
for ext in extensions:
    all_files += glob.glob(os.path.join(DATA_DIR, ext))
if not all_files:
    raise RuntimeError(f"No images found in {DATA_DIR}")

total_count = len(all_files)
print(f"Found {total_count} images.")

# 2) Create tf.data Dataset of paths & shuffle
ds_paths = tf.data.Dataset.from_tensor_slices(all_files)
ds_paths = ds_paths.shuffle(buffer_size=total_count, seed=42)

# 3) Split into train / val
train_count = int(total_count * SPLIT_RATIO)
train_paths = ds_paths.take(train_count)
val_paths   = ds_paths.skip(train_count)

# 4) Loader → (img, img) for autoencoder
def load_img(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)  # [0,1]
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])     # ensure target size
    return img, img

train_ds = train_paths.map(load_img, num_parallel_calls=AUTOTUNE)
val_ds   = val_paths.map(load_img,   num_parallel_calls=AUTOTUNE)

# 5) Define augmentation & masking (no TFA)
flip_layer     = tf.keras.layers.RandomFlip(mode="horizontal_and_vertical")
rotate_layer   = tf.keras.layers.RandomRotation(factor=0.35)
zoom_layer     = tf.keras.layers.RandomZoom(height_factor=0.2, width_factor=0.2)
contrast_layer = tf.keras.layers.RandomContrast(factor=0.2)

def apply_random_masks(x, num_masks=1, mask_size=MASK_SIZE):
    """Apply num_masks random square masks to x, avoiding the upper-left UL_REGION."""
    for _ in range(num_masks):
        mx = tf.random.uniform([], 0, IMG_SIZE - mask_size + 1, dtype=tf.int32)
        my = tf.random.uniform([], 0, IMG_SIZE - mask_size + 1, dtype=tf.int32)

        # Ensure mask does not start in upper-left region
        mx = tf.cond(
            tf.logical_and(mx < UL_REGION, my < UL_REGION),
            lambda: UL_REGION,
            lambda: mx
        )

        ys = tf.range(IMG_SIZE)[:, None]
        xs = tf.range(IMG_SIZE)[None, :]
        region = (ys >= my) & (ys < my + mask_size) & \
                 (xs >= mx) & (xs < mx + mask_size)
        region = tf.expand_dims(region, -1)  # [H,W,1]

        mask = tf.where(region, tf.zeros_like(x), tf.ones_like(x))
        x = x * mask
    return x

def augment_and_mask(x, y):
    # 1) Keras preprocessing augmentations
    x = flip_layer(x, training=True)
    x = rotate_layer(x, training=True)
    x = zoom_layer(x, training=True)
    x = contrast_layer(x, training=True)

    # 2) Color jitter + Gaussian noise
    x = tf.image.random_brightness(x, max_delta=0.2)
    x = tf.image.random_hue(x, 0.08)
    x = tf.image.random_saturation(x, 0.7, 1.3)
    noise = tf.random.normal(tf.shape(x), mean=0.0, stddev=0.02)
    x = tf.clip_by_value(x + noise, 0.0, 1.0)

    # 3) Apply multiple random masks
    x = apply_random_masks(x, num_masks=NUM_MASKS, mask_size=MASK_SIZE)

    return x, y

# 6) Create augmented dataset & concatenate with original
aug_ds = train_ds.map(augment_and_mask, num_parallel_calls=AUTOTUNE)
combined_train_ds = train_ds.concatenate(aug_ds)

# 7) Shuffle, batch, prefetch
train_ds = (combined_train_ds
            .shuffle(buffer_size=2*train_count, seed=42)
            .batch(BATCH_SIZE)
            .prefetch(AUTOTUNE))

val_ds = (val_ds
          .batch(BATCH_SIZE)
          .prefetch(AUTOTUNE))

print("✅ Training and validation datasets ready.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def visualize_samples(dataset, n=8, from_end=False):
    """
    Show first n or last n samples from a (img, img) dataset.
    """
    # Materialize to list (only for finite datasets)
    imgs = [x.numpy() for x, _ in dataset.unbatch()]
    
    if from_end:
        imgs = imgs[-n:]
    else:
        imgs = imgs[:n]
    
    plt.figure(figsize=(3*n, 3))
    for i, img in enumerate(imgs):
        plt.subplot(1, n, i + 1)
        plt.imshow(img)
        plt.axis("off")
    plt.show()

visualize_samples(train_ds, n=8, from_end=True)   # last 8
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model, callbacks, metrics

# -------------------------
# Variational Transformer Encoder
# -------------------------
class VariationalTransformerEncoder(Model):
    def __init__(self, latent_dim=64, context_n=5, ff_dim=128, num_heads=4, **kwargs):
        super().__init__(**kwargs)
        self.latent_dim = latent_dim
        self.context_n = context_n
        self.ff_dim = ff_dim
        self.pos_emb = layers.Embedding(input_dim=context_n + 1, output_dim=ff_dim)
        self.attn1 = layers.MultiHeadAttention(num_heads=num_heads, key_dim=ff_dim)
        self.attn2 = layers.MultiHeadAttention(num_heads=num_heads, key_dim=ff_dim)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim * 2, activation='relu'),
            layers.Dense(ff_dim),
        ])
        self.norm1 = layers.LayerNormalization()
        self.norm2 = layers.LayerNormalization()
        self.norm3 = layers.LayerNormalization()
        self.z_mean = layers.Dense(latent_dim)
        self.z_log_var = layers.Dense(latent_dim)

    def call(self, seq_embedded, training=False):
        x = seq_embedded
        seq_len = tf.shape(x)[1]
        positions = tf.range(seq_len)
        x += self.pos_emb(positions)
        a1 = self.attn1(x, x)
        x = self.norm1(x + a1)
        a2 = self.attn2(x, x)
        x = self.norm2(x + a2)
        f = self.ffn(x)
        x = self.norm3(x + f)
        x_pool = tf.reduce_mean(x, axis=1)
        z_mean = self.z_mean(x_pool)
        z_log_var = tf.clip_by_value(self.z_log_var(x_pool), -4.0, 4.0)
        eps = tf.random.normal(tf.shape(z_mean), dtype=z_mean.dtype)
        z = z_mean + tf.exp(0.5 * z_log_var) * eps
        return z, z_mean, z_log_var

# -------------------------
# Patch Decoder
# -------------------------
class PatchDecoder(Model):
    def __init__(self, patch_shape=(50,50,3), **kwargs):
        super().__init__(**kwargs)
        H, W, C = patch_shape
        self.H, self.W, self.C = H, W, C
        h8, w8 = (H+7)//8, (W+7)//8
        self.dense = layers.Dense(h8*w8*128, activation='relu')
        self.reshape = layers.Reshape((h8, w8, 128))
        self.deconv1 = layers.Conv2DTranspose(128, 3, strides=2, padding='same', activation='relu')
        self.deconv2 = layers.Conv2DTranspose(64, 3, strides=2, padding='same', activation='relu')
        self.deconv3 = layers.Conv2DTranspose(32, 3, strides=2, padding='same', activation='relu')
        # Final conv to get C channels
        self.output_conv = layers.Conv2D(C, 3, padding='same', activation='sigmoid')

    def call(self, z, training=False):
        x = self.dense(z)
        x = self.reshape(x)
        x = self.deconv1(x)
        x = self.deconv2(x)
        x = self.deconv3(x)
        # Now map to original channels and crop
        x = self.output_conv(x)
        return x[:, :self.H, :self.W, :]

# -------------------------
# Chunked VAE with Embedded Context
# -------------------------
class ChunkedVAE(Model):
    def __init__(self, encoder, decoder, h_splits, w_splits, lambda_reg=1.0, context_n=5, ff_dim=128):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.h_splits = h_splits
        self.w_splits = w_splits
        self.lambda_reg = lambda_reg
        self.context_n = context_n
        self.patch_embed = layers.Dense(ff_dim, activation='relu')
        self.kl_weight = tf.Variable(0.0, trainable=False)
        self.total_loss_tracker = metrics.Mean('total_loss')
        self.recon_loss_tracker = metrics.Mean('recon_loss')
        self.kl_loss_tracker = metrics.Mean('kl_loss')

    @property
    def metrics(self):
        return [self.total_loss_tracker, self.recon_loss_tracker, self.kl_loss_tracker]

    def _extract_and_embed(self, inputs):
        patches = tf.image.extract_patches(
            inputs,
            sizes=[1,self.decoder.H,self.decoder.W,1],
            strides=[1,self.decoder.H,self.decoder.W,1],
            rates=[1,1,1,1],
            padding='VALID'
        )
        B = tf.shape(inputs)[0]
        P = self.h_splits * self.w_splits
        feats = self.decoder.H * self.decoder.W * self.decoder.C
        flat = tf.reshape(patches, [B, P, feats])
        embedded = self.patch_embed(flat)
        return flat, embedded

    def _gather_context(self, embedded):
        B = tf.shape(embedded)[0]
        P = self.h_splits * self.w_splits
        idx = tf.range(P)
        # Build sliding window indices
        def ctx(i):
            start = tf.maximum(0, i-self.context_n)
            pad = self.context_n - (i-start)
            inds = idx[start:i]
            return tf.concat([tf.zeros(pad, dtype=tf.int32), inds], axis=0)
        ctx_idx = tf.stack([ctx(i) for i in range(P)], axis=0)
        # Expand to batch
        batch_idx = tf.range(B)[:,None,None]
        batch_idx = tf.tile(batch_idx, [1,P,self.context_n])
        ctx_idx = tf.tile(ctx_idx[None], [B,1,1])
        gather_idx = tf.stack([batch_idx, ctx_idx], axis=-1)
        return tf.gather_nd(embedded, gather_idx)

    def call(self, inputs, training=False):
        B = tf.shape(inputs)[0]
        C = tf.shape(inputs)[-1]
        flat, embedded = self._extract_and_embed(inputs)
        context = self._gather_context(embedded)
        # Reshape current embeddings
        current = tf.reshape(embedded, [B, self.h_splits*self.w_splits, 1, tf.shape(embedded)[-1]])
        seq = tf.concat([context, current], axis=2)
        seq_flat = tf.reshape(seq, [-1, self.context_n+1, tf.shape(embedded)[-1]])
        z, zm, zl = self.encoder(seq_flat, training=training)
        recon_p = self.decoder(z, training=training)
        # Stitch patches back to image
        recon = tf.reshape(recon_p, [B, self.h_splits, self.w_splits, self.decoder.H, self.decoder.W, C])
        recon = tf.transpose(recon, [0,1,3,2,4,5])
        return tf.reshape(recon, [B, self.h_splits*self.decoder.H, self.w_splits*self.decoder.W, C])

    def train_step(self, data):
        x = data[0] if isinstance(data, tuple) else data
        with tf.GradientTape() as tape:
            flat, embedded = self._extract_and_embed(x)
            context = self._gather_context(embedded)
            B = tf.shape(embedded)[0]
            P = self.h_splits * self.w_splits
            current = tf.reshape(embedded, [B, P, 1, tf.shape(embedded)[-1]])
            seq = tf.concat([context, current], axis=2)
            seq_flat = tf.reshape(seq, [-1, self.context_n+1, tf.shape(embedded)[-1]])
            z, zm, zl = self.encoder(seq_flat, training=True)
            preds = self.decoder(z, training=True)
            orig = tf.reshape(flat, [-1, self.decoder.H, self.decoder.W, self.decoder.C])
            orig_cast = tf.cast(orig, preds.dtype)
            recon_loss = tf.reduce_mean(tf.square(orig_cast - preds))
            kl_loss = -0.5 * tf.reduce_mean(1 + zl - tf.square(zm) - tf.exp(zl))
            loss = recon_loss + self.kl_weight * kl_loss
        grads = tape.gradient(loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {m.name: m.result() for m in self.metrics}

    def test_step(self, data):
        x = data[0] if isinstance(data, tuple) else data
        # Mirror train_step without gradients
        flat, embedded = self._extract_and_embed(x)
        context = self._gather_context(embedded)
        B = tf.shape(embedded)[0]
        P = self.h_splits * self.w_splits
        current = tf.reshape(embedded, [B, P, 1, tf.shape(embedded)[-1]])
        seq = tf.concat([context, current], axis=2)
        seq_flat = tf.reshape(seq, [-1, self.context_n+1, tf.shape(embedded)[-1]])
        z, zm, zl = self.encoder(seq_flat, training=False)
        preds = self.decoder(z, training=False)
        orig = tf.reshape(flat, [-1, self.decoder.H, self.decoder.W, self.decoder.C])
        orig_cast = tf.cast(orig, preds.dtype)
        recon_loss = tf.reduce_mean(tf.square(orig_cast - preds))
        kl_loss = -0.5 * tf.reduce_mean(1 + zl - tf.square(zm) - tf.exp(zl))
        loss = recon_loss + self.kl_weight * kl_loss
        self.total_loss_tracker.update_state(loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {m.name: m.result() for m in self.metrics}

# KL Warmup Callback
class KLWarmup(callbacks.Callback):
    def __init__(self, vae, ramp_epochs=50):
        super().__init__()
        self.vae = vae
        self.ramp_epochs = ramp_epochs
    def on_epoch_end(self, epoch, logs=None):
        w = min((epoch+1)/self.ramp_epochs, 1.0) * self.vae.lambda_reg
        self.vae.kl_weight.assign(w)
        tf.print(f"KL weight now {w}")
```

```python
LATENT_DIM = 64
CTX_SIZE = 10
encoder = VariationalTransformerEncoder(latent_dim=LATENT_DIM, context_n=CTX_SIZE)
decoder = PatchDecoder(patch_shape=(32, 32, 3))
autoencoder = ChunkedVAE(encoder, decoder, h_splits=8, w_splits=8, lambda_reg=0.01, context_n=CTX_SIZE)

autoencoder.compile(
    optimizer=tf.keras.optimizers.AdamW(1e-4),
    loss=None,   
    metrics=[],
    run_eagerly=True
)
```

```python
EPOCHS = 5000

# Callbacks
ckpt_cb = ModelCheckpoint("best_autoencoder.keras", save_best_only=True, monitor="val_total_loss", mode="min", verbose=1)
log_cb = CSVLogger("training_log.csv", append=True)
warmup_cb = KLWarmup(autoencoder, ramp_epochs=100)
earlystop_cb = EarlyStopping(
    monitor="val_total_loss",
    patience=200,
    restore_best_weights=True,
    mode='min'
)
```

```python
history = autoencoder.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[log_cb, ckpt_cb, warmup_cb, earlystop_cb]
)
```

```python
def show_reconstructions(model, dataset, n=5):
    sample_batch = next(iter(dataset))
    x_val = sample_batch[0] if isinstance(sample_batch, tuple) else sample_batch

    # Limit n to available samples
    n = min(n, x_val.shape[0])

    reconstructions = model.predict(x_val)

    x_val = np.clip(x_val, 0, 1)
    reconstructions = np.clip(reconstructions, 0, 1)

    plt.figure(figsize=(12, 5))
    for i in range(n):
        # Original (top row)
        plt.subplot(2, n, i+1)
        plt.imshow(x_val[i])
        plt.axis("off")
        if i == 0:
            plt.ylabel("Original")

        # Reconstructed (bottom row)
        plt.subplot(2, n, i+1+n)
        plt.imshow(reconstructions[i])
        plt.axis("off")
        if i == 0:
            plt.ylabel("Reconstructed")

    plt.tight_layout()
    plt.show()

show_reconstructions(autoencoder, val_ds, n=5)
```

```python
def show_random_generations(model, n=5):
    # Extract patch/grid info
    ph, pw = model.decoder.H, model.decoder.W
    hs, ws = model.h_splits, model.w_splits
    latent_dim = model.encoder.latent_dim  # <-- use latent_dim attribute

    plt.figure(figsize=(12, 5))
    
    for i in range(n):
        # Generate random latent codes for all patches
        random_latents = tf.random.normal([hs * ws, latent_dim])

        # Decode to patches
        random_patches = model.decoder(random_latents, training=False)  # shape: [P, ph, pw, C]

        # Stitch patches into a full image
        B = 1
        P = hs * ws
        C = random_patches.shape[-1]
        recon_reshaped = tf.reshape(random_patches, [B, hs, ws, ph, pw, C])
        recon_transposed = tf.transpose(recon_reshaped, perm=[0, 1, 3, 2, 4, 5])
        random_image = tf.reshape(recon_transposed, [B, hs * ph, ws * pw, C])
        
        # Plot
        plt.subplot(1, n, i+1)
        plt.imshow(np.clip(random_image[0].numpy(), 0, 1))
        plt.axis("off")
        plt.title("Random Gen")

    plt.tight_layout()
    plt.show()

show_random_generations(autoencoder, n=5)
```
